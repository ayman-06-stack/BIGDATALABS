```python
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("ColabSpark").config("spark.driver.memory", "2g").getOrCreate()
print(" Spark est configuré avec succès!")
```

```python
data = [(1, "Alice", 23), (2, "Bob", 30), (3, "Charlie", 29)]
columns = ["id", "nom", "age"]
df = spark.createDataFrame(data, columns)
```

```python
df.show()
```

```python
df.printSchema()
df.select("nom", "age").show()
df.filter(df.age > 25).show()
```

```python
df = spark.read.csv("/content/transaction.csv", header=True, inferSchema=True)
df.printSchema() # Inspectez les types de données pour vous assurer que Amount est un type numérique
```

```python
df = df.withColumnRenamed("Amount", "Transaction Amount")
# Utilisons la colonne 'Class' (qui indique la fraude 0/1) comme notre 'Transaction Type' pour l'exercice de groupement
df = df.withColumnRenamed("Class", "Transaction Type")

# Affichez le schéma pour vérifier les nouveaux noms de colonnes
df.printSchema()
df.show(5)
```

```python
df.groupBy("Transaction Type").sum("Transaction Amount").show()
```

```python
df.groupBy("Transaction Type").sum("Transaction Amount").show()
```

```python
from pyspark.sql.functions import col
```

```python
df.orderBy(col("Transaction Amount").desc()).show(5)
```
